```python
# data3: age_range, Embarked, Pclass, Sex
# 사용자 정의 함수
# ohe_trans(df, 'Embarked')
def ohe_trans(df, col):
    ohe = OneHotEncoder()
    x = ohe.fit_transform(df[col].values.reshape(-1,1)).toarray()
    tmp = []
    for i in range(df[col].unique().size):
        tmp.append(col[0]+str(i))
        
    ohe_df = pd.DataFrame(x, columns=tmp)
    
    return ohe_df
```

```python
result1 = ohe_trans(data3, 'age_range')
result2 = ohe_trans(data3, 'Embarked')
result3 = ohe_trans(data3, 'Pclass')
result4 = ohe_trans(data3, 'Sex')
```

```python
data3.drop(['age_range', 'Embarked', 'Pclass', 'Sex'], axis=1, inplace=True)
```

```python
data3 = pd.concat([data3, result1, result2, result3, result4], axis=1)
```

```python
r1 = ohe_trans(data4, 'age_range')
r2 = ohe_trans(data4, 'Embarked')
r3 = ohe_trans(data4, 'family_size')
r4 = ohe_trans(data4, 'Pclass')
r5 = ohe_trans(data4, 'Sex')
```
