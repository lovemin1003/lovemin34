#### 1. 모든 컬럼 이름을 소문자로 변경

```python
# 반복문
# 모든 컬럼을 소문자로 바꾼 후 새로운 컬럼으로 전달
new_cols = []
for col in data.columns:
    # str 타입의 메서드 : lower()
    new_cols.append(col.lower())
```

```python
data.colums = new_cols
```

#### 2. 남자를 0, 여자를 1로 가지는 sex_num 컬럼 추가

```python
data['sex'].head()
```

```python
# 반복문
new_sex = []
for i in data['sex']:
    if i == 'male':
        new_sex.append(0)
    else:
        new_sex.append(1)
```

```python
data['sex_num1'] = new_sex
```

```python
data[['sex','sex_num1']].head(3)
```

```python
# replace
data['sex_num2'] = data['sex'].replace('male', 0)
data['sex_num2'] = data['sex_num2'].replace('female', 1)
```

```python
data[['sex', 'sex_num2']].head(2)
```

```python
# map/apply - dict
# 1. dict 하드코딩
# 2. dict() & zip() & range()
# 3. enumerate()
dict_map = {'male':0, 'female':1}

data['sex'].map(dict_map)
```

#### 3. 이름 길이 값을 저장하는 name_length 컬럼 추가

```python
data['name'].head(2)
```

```python
# name_len = []
# for i in data['name']:
#     name_len.append(len(i))
```

```python
# lambda 함수는 간단한 실행문(함수가 아닌 단순한 제어문)을
# 사용할 때 주로 쓰인다.
data['name'].apply(lambda x : len(x))
```

#### 5. name에서 이름을 추출하여 last_name 컬럼 추가
- 이름 : 가장 처음에 나오는 이름
- 사용자 정의함수 사용
- lambda 함수 사용

```python
data['name'].head()
```

```python
# 슬라이싱
# str.find(t)
tmp = data['name'][0]
```

```python
tmp.find(' ')-1
```

```python
tmp[:6]
```

```python
# tmp[:(tmp.find(' ')-1)]
data['name'].apply(lambda x: x[:(x.find(' ')-1)])
```

```python
# split
data['name'].apply(lambda x: x.split(',')[0])
```

```python
data['name'].str.split(',').apply(lambda x: x[0])
```

```python
# 사용자 정의 함수
# 1. split한 결과에서 인덱스를 고정
def my_func1(x):
    return x[0]
```

```python
# 실행문
data['name'].str.split(',').apply(my_func1)
```

```python
# 2. split한 결과에서 인덱스를 고정하지 않은 경우 
```

```python
def my_func2(x, idx):
    return x[idx]
```

```python
# 실행문 
data['name'].str.split(',').apply(my_func2, idx=1)
```

```python
new_name = []
for i in data['name']:
    text_list = i.split(',')
#     print(text_list)
#     print(text_list[0])
    new_name.append(text_list[0])
```

```python
data['new_name'] = new_name
```
