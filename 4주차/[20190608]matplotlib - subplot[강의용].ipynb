### Matplotlib 객체 구조

- Figure : 한 개 이상의 Axes 객체 포함하는 가장 큰 단위의 객체 (캔버스 틀)
- Axes : 두 개 이상의 Axis 객체 포함 (캔버스 안의 하위 영역)
- Axis : 축 정보를 담고 있는 객체 


```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# 한글 폰트 설정
font_name = matplotlib.font_manager.FontProperties(fname='C:/Windows/Fonts/malgun.ttf').get_name()
matplotlib.rc('font', family = font_name)
```

#### 1. Figure 생성

- plt.figure() 함수 : figure 객체 반환
    - plt.plot() 함수를 실행하면 자동으로 figure 객체를 생성
    - figure 객체의 크기(plot 이미지의 크기)를 설정,생성하고 싶을 때 명시적으로 사용

```python
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# figure() 함수를 사용하지 않고 plot()으로 생성되는 figure 객체
# 표준정규분포 100개
data = np.random.randn(100)
plt.plot(data)
plt.show()
```

```python
# figure() 함수를 이용하여 figure 객체 크기를 쉽게 지정해서 생성 가능
# figure 크기 설정 : plt.figure(figsize=(w,h))
plt.figure(figsize=(10,10))
plt.plot(data)
plt.show()
```

#### 2. Subplot 생성

- 하나의 윈도우(Figure)안에 여러 개의 이미지(plot)을 생성해야 하는 경우
- subplot() 함수
    - plt.subplot(row, col, idx)
        - 단독 생성 : idx 값으로 axes의 위치 지정
        - 2,2,1 => 221
    - plt.subplots(row, col)
        - figure, axes 객체 생성 : 행/열 구조만 설정하여 axes 객체 생성한 후 각각의 axes 객체로 위치 지정

```python
# 1. subplot()으로 단독 생성
# 2행 1열짜리 레이아웃 생성
# 첫번째 플롯
plt.subplot(2,1,1)
```

```python
# 두번째 플롯
plt.subplot(2,1,2)
```

```python
# 한번에 모든 subplot을 화면에 띄우기 
# 현재는 틀만 그린 상태
plt.subplot(2,1,1)
plt.subplot(2,1,2)
```

```python
# subplot을 단독으로 생성하는 방법
# 구조 : 2 x 2 => 4칸짜리 레이아웃 => 서로 다른 모양의 라인 플롯
# 데이터 : random.rand(5) => 0~1 사이 난수 5개
# 순서 
# 1. 틀 생성
# 2. 값을 그래프로 그리기
# 3. 속성 지정
# 4. show()
plt.subplot(2,2,1)
plt.plot(np.random.rand(5))
plt.title('axes1')

plt.subplot(2,2,2)
plt.plot(np.random.rand(5))
plt.title('axes2')

plt.subplot(2,2,3)
plt.plot(np.random.rand(5))
plt.title('axes3')

plt.subplot(2,2,4)
plt.plot(np.random.rand(5))
plt.title('axes4')

plt.tight_layout(pad=1)
plt.show()
```

```python
# figure 객체와 여러 개의 axes 객체를 생성하는 방법
# fig, axes = plt.subplots(2, 2)
# plt.subplot(2,2,1) => 위치지정하는 법
# axes[0,0], [0,1], [1,0], [1,1]
fig, axes = plt.subplots(2,2)
axes[0,0].plot(np.random.rand(5))
axes[0,0].set_title('axes1')
axes[0,1].plot(np.random.rand(5))
axes[0,1].set_title('axes2')
axes[1,0].plot(np.random.rand(5))
axes[1,0].set_title('axes3')
axes[1,1].plot(np.random.rand(5))
axes[1,1].set_title('axes4')

plt.tight_layout(pad=1)
plt.show()
```

#### 3. x축을 공유하면서 서로 다른 y축을 가지는 plot 생성

- axes 객체를 생성하여 서로 다른 y축 적용
    - axes.twinx() : 동일한 x축을 공유하는 axes를 반환

```python
plt.plot([10,5,6,17,20], 'r-')
plt.plot([1000,3000,10000,7000,15000], 'g:')
plt.show()
```

```python
# 데이터 : 빨간실선, 녹색점선으로 구분해서 라인플롯
# plt.subplots()
fig, ax0 = plt.subplots()
ax1 = ax0.twinx()

ax0.plot([10,5,6,17,20], 'r-')
# 공통 속성(이미지 타이틀, x축 이름) : 원본객체에서 속성 반영해야함
ax0.set_title('공통 x축을 가지는 서로 다른 y축')
ax0.set_ylabel('y0')
ax0.set_xlabel('x')

ax1.plot([1000,3000,10000,7000,15000], 'g:')
ax1.set_ylabel('y1')
# ax1.set_xlabel('x')   # 안됨
# plt.xlabel('x')   # 안됨

plt.show()
```

### 연습문제

```python
# 데이터 생성
# x축으로 사용할 데이터 : 범주형 데이터(2~3)
# y축으로 사용할 데이터 : 정수 값 데이터 => 각각의 범주가 가지는 크기
# 세 반의 평균 점수
class_name = ['A','B','C']
scores = [70, 85, 90]
```

```python
plt.figure(figsize=(10,4))
```

```python
# 캔버스(figure) 객체는 (10,4)로 크기 적용
# bar plot 생성
# plt.bar(width=0.8(default)) : bar 굵기 변경
plt.figure(figsize=(4,3))
plt.bar(class_name, scores, width=0.9)
```

```python
# scatter plot
plt.figure(figsize=(4,3))
plt.scatter(class_name, scores)
```

```python
# line plot
plt.figure(figsize=(4,3))
plt.plot(class_name, scores)
# plt.plot(class_name, scores, 'o')
```

```python
# 모든 subplot을 한 번에 보기 (1행 3열)
# axes 객체 사용 X
# 1번 위치 : bar
# 2번 위치 : scatter
# 3번 위치 : line
# plt.subplot(r,c,idx) # 1,3,1 / 1,3,2 / 1,3,3
plt.subplot(1,3,1)   # plt.subplot(131)
plt.bar(class_name, scores)
# plt.title('ttitle')

plt.subplot(1,3,2)
plt.scatter(class_name, scores)

plt.subplot(1,3,3)
plt.plot(class_name, scores)

plt.tight_layout(pad=2)
```

```python
# subplot을 3행 1열 구조로 생성
# axes 객체 사용
# plt.subplots(r,c)
fig, axes = plt.subplots(3,1, figsize=(5,10))   # 그림 크기
axes[0].bar(class_name, scores)
axes[1].scatter(class_name, scores)
axes[2].plot(class_name, scores)

plt.tight_layout(pad=3)   # 여백
```
