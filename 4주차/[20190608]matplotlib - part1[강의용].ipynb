### Matplotlib

- 파이썬 기반 데이터 분석 환경에서 가장 일반적으로 사용되는 시각화 라이브러리
- 하위 라이브러리인 'pyplot'을 통해 주요 기능을 사용 
- plotly, seaborn 등과 같은 라이브러리가 최근 경쟁력을 갖추고 있음
- pandas 객체의 기본적인 시각화 도구
- plot() 메서드를 통해 데이터를 시각화
- 종류
    - 라인 플롯(line plot)
    - 바 차트(bar chart)
    - 히스토그램(histogram)

    - 스캐터 플롯(scatter plot)

### 스타일 지정

- plt.plot([x], y, [fmt], **kwargs)
    - fmt : '[color][marker][line]'
        - 색상, 마커, 선 등에 대한 포맷 옵션 설정
    
    <img src='img/color_line.png' width='400' height='600' align='left'>

```python
# 스타일 문자열 추가 : [fmt] => color+maker+line 순서대로 문자열 작성
# 'o' : 라인없이 큰점(원모양)으로 데이터 표시 => scatter plot과 유사
plt.plot([1,2,3,4], [2,6,10,7], 'o')
plt.show()
```

```python
# o- : 마커는 동그라미, 라인은 실선
plt.plot([1,2,3,4], [2,6,10,7], 'o-')
plt.show()
```

```python
# o-- : 마커는 동그라미, 선모양은 --
plt.plot([1,2,3,4], [2,6,10,7], 'o--')
plt.show()
```

```python
# ro: : 마커는 동그라미, 라인모양은 :, 라인색은 빨강
plt.plot([1,2,3,4], [2,6,10,7], 'ro:')
plt.show()
```

```python
# 기타 스타일 파라미터 종류
# color : 선색깔 / magenta
# linestyle : 선모양 / -.
# marker : 마커모양 / ^
# markeredgecolor : 마커선색깔 / r
# markerfacecolor : 마커색깔 / orange
# markersize : 마커크기 / 20
plt.plot([1,2,3,4], [2,6,10,7], color='magenta', linestyle='-.', marker='^',
         markeredgecolor = 'r', markerfacecolor = 'cyan', markersize=20)
plt.show()
```

```python
# 0 이상 10 미만의 범위에서 0.5 씩 증가하는 실수 생성
import numpy as np

data = np.arange(0, 10, 0.5)
data
len(data)
```

```python
# 데이터셋 1 : x,y 데이터가 동일(y=x), 파란색 -- 라인 플롯
plt.plot(data, data, 'b--')
plt.show()
```

```python
# 데이터셋 2 : y = x**2, 빨간색 네모 마커 선 없이
plt.plot(data, data**2, 'rs')
plt.show()
```

```python
# 데이터셋 3 : y=x**3, 녹색 삼각형 마커 선 없음
plt.plot(data, data**3, 'g^')
plt.show()
```

```python
# 하나의 플롯에 세 개의 그래프 시각화 1
# plot을 동일한 셀에 세번 띄우기
plt.plot(data, data, 'b--')
plt.plot(data, data**2, 'rs')
plt.plot(data, data**3, 'g^')
plt.show()
```

```python
# 하나의 플롯에 세 개의 그래프 시각화 2
# 하나의 plot에 데이터를 한번에 전달
# plt.plot(x1, y1, fmt1, x2, y2, fmt2, x3, y3, fmt3)
plt.plot(data, data, 'b--', data, data**2, 'rs', data, data**3, 'g^')
plt.show()
```

#### matplotlib.pyplot.scatter 

- 데이터가 흩어져 있는 정도를 시각화 => 데이터의 분산 정도
- 데이터 간의 상관관계를 파악할 수 있음
- plt.scatter(x, y, s=None, c=None, data=None)
    - s : 마커 사이즈
    - c : 마커 색상
    - data : dictionary 자료형을 사용하여 데이터를 전달하면 keyword string을 통해 데이터 사용 가능

```python
# 정규 분포 확률 밀도에서 100개의 표본 추출
# random.normal()
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
plt.scatter(x, y)
plt.show()
```

```python
# 딕셔너리 타입으로 데이터 전달
# 샘플 의미 : 데이터마다 색상, 마커 크기가 다르게 표현되는 시각화를 위한 샘플
# x: x값 => 0 이상 50미만의 arange
# c: 마커색상 => 0이상 50미만 정수 중 50개 난수
# r: 정규분포에서 추출한 난수 50 
# y: y값 => x 데이터 + 10*정규분포에서 추출한 난수 50(r)
# s: 마커사이즈 => r의 절대값 * 100
data = {'x' : np.arange(0,50),
       'c' : np.random.randint(0,50,50),
       'r' : np.random.randn(50)}

data['y'] = data['x'] + 10*data['r']
data['s'] = np.abs(data['r'])*100
```

```python
# 데이터 확인
data
```

```python
# 딕셔너리 데이터 타입으로 플롯 데이터 전달
# plt.scatter(x='k1',y='k2',s='k3',c='k4',data=dict)
plt.scatter(x='x', y='y', s='s', c='c', data=data)
plt.show()
```

### Bar chart (막대 그래프)

- 여러 값의 상대적인 크기를 비교하기에 적합한 표현법이며 원형 차트보다 더욱 일반적으로 사용됨
- plt.bar(x, y)
    - x 데이터가 카테고리 값인 경우에는 bar, barh 로 바 차트 시각화 표현
    - 가로 방향으로 바 차트를 그리려면 barh 

```python
# 범주형 데이터에 대한 bar graph 기본
# plt.(x, y)
x = ['가', '나', '다']
y = [5, 10, 6]
plt.bar(x, y)
```

```python
# 제목(plt.title), ticks, 축제목(plt.xlabel, ylabel) 설정
# ticks : 축에 있는 값의 이름
# xticks : x축 값 이름을 변경하는데 사용
# plt.xticks(원본값, 바꿀값)
x = ['가', '나', '다']
y = [5, 10, 6]
plt.bar(x, y)
plt.xticks(x, ['X', 'Y', 'Z'])
plt.show()
```

```python
# 각각의 bar 마다 색상 변경
# plt.bar() 객체 별로 set_color()
bar_list = plt.bar(x, y)
bar_list[0].set_color('r')
# bar_list[1].set_color('g')
plt.show()
```

### Pie chart
- 범주형 데이터
- 전체 데이터에서 각 부분이 차지하는 비율을 시각적으로 이해하기 쉽게 표현
- plt.pie(counts, labels=, colors=, autopct=, shadow=T/F)
    - counts : 각 라벨의 값(개수, 양...)
    - label : 데이터의 카테고리명
    - color : 라벨별 색상
    - autopct : 비율 출력 포맷
    - shadow : 음영 지정

```python
# 라벨, 값, 색상
# 라벨 : 소고기, 닭고기, 돼지고기, 오리고기
# 값 : 20, 35, 54, 18
# 색상 : yellowgreen, gold, lightskyblue, lightcoral
plt.pie([20, 35, 180, 18], labels=['소고기', '닭고기', '돼지고기', '오리고기'],
       colors=['yellowgreen', 'gold', 'lightskyblue', 'lightcoral'],
       autopct='%0.1f%%', shadow=True)
plt.axis('equal')  # 정원 모양으로 설정할 수 있음
plt.show()
```

### histogram(히스토그램)

- 데이터 분포의 특징, 이상치 여부를 확인하는데 용이
    - 데이터의 분포 
        - 데이터 값의 전반적으로 퍼져있는 형태
        - 데이터의 수가 많을 때 데이터를 파악하기 위해 어떤 값 근처에 어떤 모양으로 모여 있는지 이해하기 쉬움
- 데이터에 이상치가 존재하는 경우 비정상적인 히스토그램이 생성되며 이상치를 제거해주는 작업이 필요
    - 예) 전반적인 데이터 분포가 0 ~ 1 사이에 있는데 한 개의 데이터가 100인 경우 정상적인 히스토그램이 그려지지 않음.

- 히스토그램
    - 자료 값이 가질 수 있는 범위를 몇 개의 구간으로 나누고 각 구간에 해당하는 값의 숫자 혹은 상대적 빈도를 계산하여 시각화
    - plt.hist(data, bins)
        - 일반적인 경우 히스토그램의 구간을 나눌 때 동일한 간격으로 나누기 때문에 구간(bins)의 개수가 중요
        - bins
            - 히스토그램 구간 수를 나타내는 값
            - 구간 수가 너무 작으면(=간격이 너무 넓으면) 세부 분포 파악이 어렵고, 구간 수가 너무 많으면 전체적인 분포 파악이 어려움

```python
# 정규분포에서 무작위로 1000개 데이터 추출
data = np.random.randn(1000)
plt.hist(data,bins=30)  # None, 30, 50, 100, 200 ...
plt.show()
```

### Box-plot
- box-and-whisker plot
- 최댓값, 최솟값, 사분위수(Q1, Q2, Q3)를 사용하여 자료의 측정값들의 분포와 극단값에 대해 파악
    - IQR(Inter Quartile Range) : Q3 - Q1
    - 최대값 : Q1 + 1.5 x IQR 범위 안에 있는 최대값
    - 최소값 : Q3 - 1.5 x IQR 범위 안에 있는 최대값
    - 박스(Box) : Q1 ~ Q3의 값
    - n개의 데이터가 있을 때 사분위수
        - 중앙값(Q2) : (n + 1) / 2
        - 사분위수
            - 자료를 크기순으로 나열하여 사등분하는 척도
            - Q2위치((n + 1) / 2)를 기준으로 작은 쪽을 아래사분위수(Q1 = 아래사분위 범위의 중앙값), 큰 쪽을 위사분위수(Q3 = 위사분위 범위의 중앙값)
            - 데이터 개수가 홀수일 때는 앞뒤 값의 평균을 사용
- 두 개 이상의 자료 집단을 비교할 때 유용
- 해당 범위에 벗어나는 값은 이상치로 판정하여 표기
